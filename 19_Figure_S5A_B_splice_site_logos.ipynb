```python
import pandas as pd
import logomaker
import matplotlib.pyplot as plt
```

```python
whip= pd.read_csv('whippet_CE_gene_name_exon_number_maxent.csv')

first_block_skip = pd.read_csv('first_block_exon.moreSKIPPED.csv')
last_block_skip = pd.read_csv('last_block_exon.moreSKIPPED.csv')

single_reg_exons = pd.read_csv('other_regulated_exons.csv')

sig_whip = pd.read_csv('sig_whip_CE_prob0.9_dPSI0.1.csv')
```

```python
# Create copies of whip based on Coord matching
fbe = whip[whip['Coord'].isin(first_block_skip['Coord'])].copy()
lbe = whip[whip['Coord'].isin(last_block_skip['Coord'])].copy()

single = whip[whip['Coord'].isin(single_reg_exons['Coord'])].copy()

# Create maxent_non_reg: whip rows that are NOT in sig_whip
non_reg = whip[~whip['Coord'].isin(sig_whip['Coord'])].copy()
```

```python
# Convert sequences into a count matrix
counts_matrix = logomaker.alignment_to_matrix(lbe['5ss_seq'], to_type='counts')

# Convert counts to probabilities
prob_matrix = logomaker.transform_matrix(counts_matrix, from_type='counts', to_type='probability')

# Define custom colors for A, C, G, T
custom_colors = {
    'A': "#A8D5E2",
    'C': "#F9A620",
    'G': "#FFD449",
    'T': "#548C2F"
}

# Create and plot sequence logo with custom colors
plt.figure(figsize=(10, 4))
logo = logomaker.Logo(prob_matrix, color_scheme=custom_colors)
logo.ax.set_ylabel('Probability')
logo.ax.set_xlabel('Position')
plt.title("5'ss last block exon")
plt.savefig('logo_5ss_last_block_exon.pdf')
plt.show()
```
